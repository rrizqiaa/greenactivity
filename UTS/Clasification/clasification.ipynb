```python
import pandas as pd

# Load the dataset
file_path = 'DATA (1).csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of grades
plt.figure(figsize=(8, 5))
sns.histplot(data['GRADE'], bins=7, kde=True, color='skyblue')
plt.title('Distribution of Grades')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.show()

# Compute the correlation matrix
correlation_matrix = data.drop(columns=['STUDENT ID', 'COURSE ID']).corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Attributes and Grade')
plt.show()
import matplotlib.pyplot as plt
import seaborn as sns


```
