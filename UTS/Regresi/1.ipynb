```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'RegresiUTSTelkom.csv'  # Replace with your file's path
data = pd.read_csv(file_path)

# Display basic information about the dataset
print("Dataset Info:")
print(data.info())

print("\nMissing Values:")
print(data.isnull().sum())

print("\nBasic Statistical Summary:")
print(data.describe())

# EDA and Data Visualization
# 1. Distribution of numeric columns (Histograms)
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
data[numeric_cols].hist(figsize=(15, 10), bins=30, color='skyblue', edgecolor='black')
plt.suptitle("Distribution of Numeric Columns", fontsize=16)
plt.show()

# 2. Boxplots to check for outliers
plt.figure(figsize=(15, 8))
sns.boxplot(data=data[numeric_cols], orient='h', palette='Set2')
plt.title("Boxplots of Numeric Columns", fontsize=16)
plt.show()

# 3. Correlation heatmap
plt.figure(figsize=(12, 8))
corr_matrix = data[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap", fontsize=16)
plt.show()

# 4. Pairplot to visualize relationships between numeric features
sns.pairplot(data[numeric_cols], diag_kind="kde", corner=True)
plt.suptitle("Pairplot of Numeric Features", y=1.02, fontsize=16)
plt.show()

# Insights
print("\n--- Insights ---")
print("1. The histogram shows the distribution of each numeric column.")
print("2. Boxplots highlight the presence of potential outliers.")
print("3. The heatmap reveals correlations between numeric features.")
print("4. Pairplot provides detailed relationships and distribution.")

# Save cleaned data (if applicable)
# Uncomment this if any data cleaning is required
# data
```
