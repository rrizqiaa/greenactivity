Eksplorasi Data:

Menampilkan beberapa baris awal data dengan df.head().
Menampilkan informasi dataset (struktur data, tipe kolom, dan jumlah nilai non-null) menggunakan df.info().
Menampilkan statistik deskriptif dataset (seperti rata-rata, standar deviasi, min, dan max) menggunakan df.describe().
Visualisasi Distribusi Kolom:

Membuat histogram untuk kolom "R", "G", "B", dan "Grey" (menggunakan indeks kolom [2:] untuk memilih kolom tersebut). Histogram menampilkan distribusi frekuensi nilai untuk setiap kolom.
Scatter Plot RGB terhadap Grey:

Membuat scatter plot untuk menampilkan hubungan antara nilai Grey (sumbu x) dan nilai R, G, B (sumbu y). Setiap warna RGB diwakili oleh titik dengan warna sesuai.
Korelasi Antar Fitur:

Menghitung matriks korelasi antar kolom menggunakan df.corr().
Menampilkan matriks korelasi dalam bentuk heatmap untuk memvisualisasikan hubungan linear antar fitur.
Tujuan Utama:
Kode ini membantu dalam memahami data sebelum melanjutkan ke analisis lebih dalam atau membangun model. Aktivitas ini mencakup:

Memahami distribusi data (statistik deskriptif dan histogram).
Menilai hubungan antar fitur (scatter plot dan matriks korelasi).
Mengidentifikasi pola atau outlier yang mungkin penting.

```python
# Import library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset
# Ganti 'dataset.csv' dengan path file Anda
df = pd.read_csv("TestPad_PCB_XYRGB_V2.csv")

# Menampilkan beberapa baris awal dataset
print("Beberapa baris awal dataset:")
print(df.head())

# Menampilkan informasi dataset
print("\nInformasi dataset:")
print(df.info())

# Statistik deskriptif
print("\nStatistik deskriptif:")
print(df.describe())

# Visualisasi distribusi nilai pada setiap kolom
columns = df.columns[2:]  # Mengambil kolom R, G, B, dan Grey
for col in columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, color="blue", bins=30)
    plt.title(f"Distribusi {col}")
    plt.xlabel(col)
    plt.ylabel("Frekuensi")
    plt.grid(True)
    plt.show()

# Visualisasi scatter plot R, G, B terhadap Grey
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Grey", y="R", data=df, label="R", color="red")
sns.scatterplot(x="Grey", y="G", data=df, label="G", color="green")
sns.scatterplot(x="Grey", y="B", data=df, label="B", color="blue")
plt.title("Scatter plot Grey vs R, G, B")
plt.xlabel("Grey")
plt.ylabel("Nilai RGB")
plt.legend()
plt.grid(True)
plt.show()

# Korelasi antar fitur
plt.figure(figsize=(8, 6))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matriks Korelasi")
plt.show()
```
